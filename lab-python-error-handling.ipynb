def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
def calculate_total_price(products):
    total_price = 0

    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                total_price += price
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}. Please enter a valid price.")

    return total_price
````

````python
def get_customer_orders(inventory):
    orders = []

    # Nhập số lượng đơn hàng
    valid_number = False
    while not valid_number:
        try:
            num_orders = int(input("How many products do you want to order? "))
            if num_orders <= 0:
                raise ValueError("Number of orders must be positive.")
            valid_number = True
        except ValueError as error:
            print(f"Error: {error}")

    # Nhập từng sản phẩm
    for i in range(num_orders):
        valid_product = False
        while not valid_product:
            try:
                product = input(f"Enter product name #{i+1}: ")

                if product not in inventory:
                    raise KeyError("Product does not exist.")

                if inventory[product] <= 0:
                    raise ValueError("Product is out of stock.")

                orders.append(product)
                inventory[product] -= 1
                valid_product = True

            except KeyError as error:
                print(f"Error: {error} Please enter a valid product name.")
            except ValueError as error:
                print(f"Error: {error}")

    return orders
````
